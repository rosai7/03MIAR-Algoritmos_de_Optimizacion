### Respuesta

La función objetivo $f(x)$ se define como el valor escalar resultante de la evaluación aritmética de la combinación de cifras y operadores elegida.

Aunque el objetivo principal del seminario es la **exploración exhaustiva (enumeración)** para identificar todos los posibles valores enteros, desde el punto de vista de la optimización el problema se desglosa en dos vertientes:
1. **Maximización:** Para hallar el límite superior del conjunto de resultados ($\max f(x)$).
2. **Minimización:** Para hallar el límite inferior del conjunto de resultados ($\min f(x)$).

Existe una restricción "dura" de validez: la solución solo se contabiliza si el resultado pertenece al conjunto de los números enteros ($f(x) \in \mathbb{Z}$), descartando cualquier combinación que genere decimales no nulos.

````
def funcion_objetivo(cifras, operadores):
    # En un entorno de optimización, aquí se implementaría la jerarquía
    # o el cálculo acumulado. Para la definición del modelo:
    res = cifras[0]
    for i in range(4):
        op = operadores[i]
        nxt = cifras[i+1]
        if op == '+': res += nxt
        elif op == '-': res -= nxt
        elif op == '*': res *= nxt
        elif op == '/': res /= nxt
    return res

# El objetivo del problema es encontrar:
# max f(x) y min f(x) sujeto a f(x) ∈ ℤ
````

## 4. Fuerza bruta

Diseña un algoritmo para resolver el problema por fuerza bruta

### Respuesta

Para el algoritmo por fuerza bruta se utilizarán 2 listas, la primera con las cifras disponibles para realizar los cálculos y la segunda con los operadores disponibles. Los valores obtenidos se irán guardando en un set.

En primer lugar, se generan todas las permutaciones posibles de ambas listas y se guardas en una lista de listas, por ejemplo, para las pemutaciones de las cifras tendremos: [[1, 2, 3, 4, 5], [1, 2, 3, 4, 6], ...]

Una vez obtenidas, pasamos a recorrerlas todas y calcular todos los valores posibles construyendo la expresión matemática:
> {Número 1}{Operador 1}{N2}{O2}{N3}{O3}{N4}{O4}{N5}.

Como en el ejemplo: 4+2-6/3*1

Una vez construida utilizamos ```eval()``` para obtener el resultado de la expresión. En caso de que no se haya producido una división por cero y que el resultado sea un entero, se añade al set de valores obtenidos.

Para acabar, respondemos a las preguntas que plantea el enunciado:

* ¿Qué valor máximo y mínimo se pueden obtener según las condiciones del problema?

* ¿Es posible encontrar todos los valores enteros posibles entre dicho mínimo y máximo ?

Para ello, ordenamos los valores de menor a mayor y tomamos el primer (mínimo) y el último elemento (máximo) y comprobamos si el número de valores obtenidos es igual a la diferencia entre el máximo y el mínimo más 1, lo que significaría que se han obtenido todos los valores intermedios.

La función devuelve los valores mínimo, máximo y un booleano que es ```True``` en caso de que la respuesta a la segunda pregunta sea afirmativa y ```False``` en caso contrario.




````
def generar_permutaciones(elementos:list, longitud:int) -> list[list]:
    """
    Genera permutaciones de una longitud dada a partir de una lista de elementos,
    sin repetir elementos (variaciones sin repetición).

    Args:
        elementos(list): Lista de elementos a permutar
        longitud(list): Número de elementos en la permutación

    Returns:
        list[list]: listado de permutaciones posibles
    """
    if longitud == 0:
        return [[]]

    resultado = []
    for i in range(len(elementos)):
        # Tomamos el elemento actual
        actual = elementos[i]
        # Creamos una lista con los elementos restantes
        resto = elementos[:i] + elementos[i+1:]

        # Llamada recursiva para completar la longitud deseada
        for p in generar_permutaciones(resto, longitud - 1):
            resultado.append([actual] + p)

    return resultado

def algoritmo_fuerza_bruta(lista_numeros:list[int]) -> tuple[int, int, bool]:
    """
    Calcula el valor máximo, mínimo y si se pueden calcular todos los números
    naturales intermedios o no de un listado de números naturales combinándolos
    con los 4 operadores algebraícos (+,-,*,/) por algoritmo de fuerza bruta.
    Args:
        lista_numeros (list(int)): listado de números sobre el que aplicar el algoritmo
    Returns:
        Tuple:
        * int: valor mínimo calculado
        * int: valor máximo calculado
        * bool: True si se pueden calcular todos los valores intermedios o False en caso contrario
    """
    print(f"Procesando lista: {lista_numeros}")
    if len(lista_numeros) < 5:
        raise Exception ("La lista debe tener al menos 5 números para usar los 4 operadores.")

    operadores = ['+', '-', '*', '/']

    resultados_enteros = set()
    valor_minimo = None
    valor_maximo = None

    # 1. Generar permutaciones de operadores
    perms_operadores = generar_permutaciones(operadores, 4)

    # 2. Generar permutaciones de lista_numeros
    perms_numeros = generar_permutaciones(lista_numeros, 5)

    total_combinaciones = len(perms_numeros) * len(perms_operadores)
    print(f"Evaluando {total_combinaciones} expresiones")

    # 3. Recorrer todas las permutaciones
    for nums in perms_numeros:
        for ops in perms_operadores:
            # Construcción de la expresión: N1 O1 N2 O2 N3 O3 N4 O4 N5
            # Ejemplo: "4+2-6/3*1"
            expresion = (f"{nums[0]}{ops[0]}{nums[1]}{ops[1]}{nums[2]}{ops[2]}{nums[3]}{ops[3]}{nums[4]}")

            try:
                # Evaluamos la expresión obtenida
                val = eval(expresion)

                # Verificamos si es entero o muy cercano a entero
                if abs(val - round(val)) < 1e-9:
                    resultados_enteros.add(int(round(val)))
                    if not valor_minimo and not valor_maximo:
                        valor_minimo = int(round(val))
                        valor_maximo = int(round(val))
                    elif int(round(val)) < valor_minimo:
                        valor_minimo = int(round(val))
                    elif int(round(val)) > valor_maximo:
                        valor_maximo = int(round(val))

            except ZeroDivisionError:
                continue

    # 4. Respondemos preguntas
    print(f"\n--- Resultados ---")
    if not resultados_enteros:
        print("No se encontraron soluciones.")
        return None, None, None
    # Buscamos minimo y maximo
    print(f"Valor Mínimo obtenido: {valor_minimo}")
    print(f"Valor Máximo obtenido: {valor_maximo}")

    # Calculamos el número de valores que existe entre el máximo y el mínimo
    n_valores_total = valor_maximo - valor_minimo + 1

    # Comprobamos si es igual que la longitud de los valores obtenidos
    if n_valores_total == len(resultados_enteros):
        todos_los_valores = True
        print(f"SÍ, es posible encontrar TODOS los valores enteros entre {valor_minimo} y {valor_maximo}.")
    else:
        todos_los_valores = False
        print(f"NO es posible encontrar todos los valores.")
        print(f"Faltan {n_valores_total - len(resultados_enteros)} números para completar la serie.")

    return valor_minimo, valor_maximo, todos_los_valores


# Ejecutamos el algoritmo
valores = [1, 2, 3, 4, 5, 6, 7, 8, 9]
algoritmo_fuerza_bruta(valores)
````

### Complejidad

Calcula la complejidad del algoritmo por fuerza bruta

#### Respuesta

Para analizar la complejidad del algoritmo de fuerza bruta, debemos evaluar cómo crecen los recursos (tiempo y memoria) en función de la cantidad de números disponibles en la lista de entrada, a la que llamaremos $n$:

* **Complejidad Temporal:** Es **O(n^5)**. Estamos calculando todas las variaciones posibles de 5 elementos sobre el total de números ($n$) y cruzándolas con las combinaciones de operadores. Además, al usar `eval()`, obligamos a Python a leer strings y convertirlo a matemáticas en cada iteración, lo cual hace que el algoritmo sea bastante lento.
* **Complejidad Espacial (Memoria):** También es **O(n^5)**. Este es el mayor problema del algoritmo: antes de calcular nada, generamos y guardamos de golpe **todas** las permutaciones en listas gigantes en la memoria RAM.

````
def P(n,k):
  return math.factorial(n) // math.factorial(n - k)

def complejidad_fuerza_bruta(n):
    if n < 5:
        return None
    num_perms = P(n, 5)
    op_perms = math.factorial(4)
    total_expr = num_perms * op_perms
    return {
        "P(n,5)": num_perms,
        "4!": op_perms,
        "total": total_expr,
        "complejidad temporal": "Θ(n^5)",
        "complejidad espacial": "Θ(n^5) (por almacenar permutaciones + set)"
    }

# Ejemplo con n=9
complejidad_fuerza_bruta(9)
````

## Optimización del Algoritmo

(*)Diseña un algoritmo que mejore la complejidad del algortimo por fuerza bruta. Argumenta porque crees que mejora el algoritmo por fuerza bruta

### Respuesta

Para el algoritmo optimizado se han utilizado las técnicas de vuelta atrás junto con poda.

En primer lugar, se ha definido la clase AlgoritmoOptimizado() para poder hacer uso de los atributos valor_mínimo y valor_máximo como valor global en las invocaciones recursivas, de forma que estos valores sean compartidos y actualizables desde cualquier invocación de los métodos de la instacia.

En el método __init__(), la clase toma como argumento la lista de números disponibles para construir la expresión matemática:
> {Número 1}{Operador 1}{N2}{O2}{N3}{O3}{N4}{O4}{N5}

Valida que esta contenga al menos 5 elementos e inicializa los atributos, construyendo las 4 listas indicadas en el apartado de "2. Modelo para el espacio de soluciones". También se inicializan los atributos que contendrán el resultado de aplicar el algoritmo.

A contuación, se implementan 3 métodos:
* `aplicar()`: Aplica el algoritmo y responde a las preguntas indicadas en los apartados anteriores.
* `_calcular_valores()`: Es el método interno que implementa el árbol de expansión y los mecanismo de recursividad hasta construir la expresión matemática completa. Aplica la vuelta atrás para explorar todas las variaciones sin necesidad de recalcular las expresiones completas.
* `_es_prometedora()`: Es el método interno que implementa la función de poda. En este caso sólo se descartan los nodos que produzcan una división por cero, ya que se quiere conocer todos los valores que se pueden calcular con la expresión matemática.

El método `aplicar()` construye la primera etapa del árbol de expansión, que contendrá un nodo por cada elemento de la lista de números recibida. Este método invoca a `_calcular_valores()`, marca el elemento correspondiente como "usado" y lo libera cuando va a pasar al siguiente nodo del primer nivel. Una vez termina de recorrer la lista de números completa, responde a las preguntas planteadas en el enunciado, del mismo modo que se hizo en el algoritmo de fuerza bruta y devuelve los valores mínimo, máximo y un booleano que es ```True``` en caso de que la respuesta a la segunda pregunta sea afirmativa y ```False``` en caso contrario.

Por otro lado, el método `_calcular_valores` contiene la parte principal del algoritmo utiliza los parámetros:

* etapa (int): Etapa de la recursión, para saber cuántos números hemos colocado o lo que es lo mismo, qué profundidad tenemos en el arbol de expansión.
* valor_actual (float): Valor acumulado de la expresión.
* valor_anterior (float): Para ajustar la jeraraquía de operaciones (* y /).

Lo primero que hace es comprobar si se ha llegado a la última etapa, la 5, en cuyo caso comprueba si el valor calculado es entero y lo almacena en el set y actualiza los valores máximos y mínimos.

Si aun faltan etapas, continua construyendo la expresión, haciendo uso de los operadores y números no usados en la rama. Para respetar la jerarquía de las operaciones combinadas, en caso de que el operador que se esté añadiendo sea * o /, se restamos el valor anterior del valor actual. El nuevo "último término calculado" pasa a ser el producto o la división, que se suma a esta diferencia. De este modo evitamos calcular `(A+B)/C` calculando `A+B-B+(B/C)`. Cuando se completan todas las posibilidades de una etapa se vuelve atrás y se desmarcan los valores usados.


````
class AlgoritmoOptimizado():
    """
    Calcula el valor máximo, mínimo y si se pueden calcular todos los números
    naturales intermedios o no de un listado de números naturales combinándolos
    con los 4 operadores algebraícos (+,-,*,/) construyendo un arbol de expansión
    para aplicar los algoritmos de vuelta atrás y ramificación y poda
    """

    def __init__(self, lista_numeros):
        # Validación inicial de número de cifras recibidas, ya que se necesitan 5.
        if len(lista_numeros) < 5:
            raise Exception ("La lista debe tener al menos 5 números para usar los 4 operadores.")

        # Inicialización de atributos
        self.lista_numeros = lista_numeros
        self.len_numeros = len(lista_numeros)
        self.operadores = ['+', '-', '*', '/']

        # Almacenan True o False en función de si se ha usado el número/operador de la misma posición
        self.usado_num = [False] * self.len_numeros
        self.usado_op = [False] * 4

        # Resultados
        self.resultados_enteros = set() # Todos los números calculados
        self.valor_minimo = None # Para ir actualizando según se añaden al set
        self.valor_maximo = None # Para ir actualizando según se añaden al set

    def aplicar(self) -> tuple[int, int, bool]:
        """
        Aplica el algoritmo optimizado y devuelve los resultados.
        Returns:
            Tuple:
                * int: valor mínimo calculado
                * int: valor máximo calculado
                * bool: True si se pueden calcular todos los valores intermedios o False en caso contrario
        """
        print(f"Procesando lista: {self.lista_numeros}")

        # Etapa 1: Iterativamente, se alimenta la función recursiva con cada uno de las cifras posibles.
        for i in range(self.len_numeros):
            self.usado_num[i] = True
            val = float(self.lista_numeros[i])
            self._calcular_valores(1, val, val)

            # Vuelta atrás
            # Cuando se pasa al siguiente valor como primer elemento esta cifra se libera.
            self.usado_num[i] = False

        # Respondemos preguntas:
        if not self.resultados_enteros:
            print("No se encontraron soluciones enteras.")
            return None, None, None

        # Mostramos minimo y maximo
        print(f"Valor Mínimo obtenido: {self.valor_minimo}")
        print(f"Valor Máximo obtenido: {self.valor_maximo}")

        # Calculamos el número de valores que existe entre el máximo y el mínimo
        n_valores_total = self.valor_maximo - self.valor_minimo + 1

        # Comprobamos si es igual que la longitud de los valores obtenidos
        if n_valores_total == len(self.resultados_enteros):
            todos_los_valores = True
            print(f"SÍ, es posible encontrar TODOS los valores enteros entre {self.valor_minimo} y {self.valor_maximo}.")
        else:
            todos_los_valores = False
            print(f"NO es posible encontrar todos los valores.")
            print(f"Faltan {n_valores_total - len(self.resultados_enteros)} números para completar la serie.")

        return self.valor_minimo, self.valor_maximo, todos_los_valores

    # Método que calcula los valores posibles
    def _calcular_valores(self, etapa:int, valor_actual:float, valor_anterior:float) -> None:
        """
        Calcula los valores posibles construyendo un árbol de expansión y aplicando vuelta atrás.
        Args:
            etapa (int): Etapa de la recursión (cuántos números hemos colocado)
            valor_actual (float): Valor acumulado de la expresión
            valor_anterior (float): Para ajustar la jeraraquía de operaciones (* y /)
        Returns:
            None
        """
        # Si se han colocado los 5 números, se añade el valor al set
        if etapa == 5:
            # Verificamos si es entero (con tolerancia float)
            if abs(valor_actual - round(valor_actual)) < 1e-9:
                # Añadimos al set
                self.resultados_enteros.add(int(round(valor_actual)))
                # Actualizamos máximo y mínimo
                if not self.valor_minimo and not self.valor_maximo:
                    self.valor_minimo = int(round(valor_actual))
                    self.valor_maximo = int(round(valor_actual))
                elif int(round(valor_actual)) < self.valor_minimo:
                    self.valor_minimo = int(round(valor_actual))
                elif int(round(valor_actual)) > self.valor_maximo:
                    self.valor_maximo = int(round(valor_actual))
            return

        # Iteramos por los operadores disponibles
        for i_op, op in enumerate(self.operadores):
            if not self.usado_op[i_op]:

                # Iteramos por los números disponibles (Hijos del nodo actual)
                for i_num in range(self.len_numeros):
                    if not self.usado_num[i_num]:

                        num = self.lista_numeros[i_num]

                        if self._es_prometedora(op, num):

                            # Marcar como usado (Avanzar etapa)
                            self.usado_op[i_op] = True
                            self.usado_num[i_num] = True

                            # Añadir términos a la operación combinada.
                            # En caso de que el nuevo operador sea * o / y tenga prioridad algebráica
                            # restamos el valor anterior del valor actual. El nuevo último término
                            # pasa a ser el producto o la división, que se suma a esta diferencia.
                            nuevo_val = 0
                            nuevo_ult = 0

                            if op == '+':
                                nuevo_val = valor_actual + num
                                nuevo_ult = num
                            elif op == '-':
                                nuevo_val = valor_actual - num
                                nuevo_ult = -num
                            elif op == '*':
                                # Pasamos de (A + B) * C a A + (B*C)
                                nuevo_ult = valor_anterior * num
                                nuevo_val = (valor_actual - valor_anterior) + nuevo_ult
                            elif op == '/':
                                # Pasamos de (A + B) / C a A + (B/C)
                                nuevo_ult = valor_anterior / num
                                nuevo_val = (valor_actual - valor_anterior) + nuevo_ult

                            # Recursión (Siguiente etapa) En el siguiente nodo, el valor actual
                            # será el nuevo valor calculado en este
                            self._calcular_valores(etapa + 1, nuevo_val, nuevo_ult)

                            # Desmarcamos como usado (Vuelta atrás)
                            # Pasamos al siguiente valor del nivel anterior, por lo que esta cifra se libera.
                            self.usado_num[i_num] = False
                            self.usado_op[i_op] = False

    # Método de Poda
    @staticmethod
    def _es_prometedora(op:str, num:int) -> bool:
        """
        Identifica si la próxima operación es una división por 0 para, en tal caso
        descartarla. Devuelve `True` si la operación "es prometedora" y `False` en
        caso contrario.
        Args:
            op (str): operador a utilizar
            num (int): segundo componente de la operación
        Returns:
            bool: `True` si no es una división por 0 y `False` si lo es.
        """
        if op == '/' and num == 0:
            return False
        return True
# Ejecutamos el algoritmo
valores = [1,2,3,4,5,6,7,8,9]

AlgoritmoOptimizado(valores).aplicar()
````

### Complejidad

(*)Calcula la complejidad del algoritmo

#### Respuesta

El algoritmo optimizado utiliza la técnica de **Backtracking** aplicando evaluaciones matemáticas rapidamente y respetando la jerarquía de operaciones sin necesidad de parsear texto.

* **Complejidad Temporal:** En teoría sigue siendo **O(n^5)**, porque en el peor de los casos seguimos teniendo que explorar todas las combinaciones para encontrar los números. Sin embargo, el tiempo real baja muchísimo porque ya no usamos `eval()` (hacemos los cálculos matemáticos directamente) y porque podamos el árbol (si vemos que vamos a dividir por cero, paramos y saltamos a otra combinación).
* **Complejidad Espacial (Memoria):** Baja drásticamente a **O(n)**. Gracias al *backtracking*, ya no guardamos todas las combinaciones de golpe. Solo guardamos en la memoria el camino exacto que estamos evaluando en ese instante (un máximo de 5 números) y un par de listas pequeñas para saber qué números hemos usado ya.

````
def complejidad_espacial_comparacion(n):
    """
    Compara cuántos elementos simultáneos guarda en memoria cada algoritmo
    antes de evaluar los resultados.
    """
    if n < 5:
        return "Necesitamos al menos 5 números."

    # --- FUERZA BRUTA ---
    combinaciones_numeros = math.factorial(n) / math.factorial(n - 5)
    combinaciones_operadores = 24
    memoria_fuerza_bruta = (combinaciones_numeros * 5) + (combinaciones_operadores * 4)

    # --- OPTIMIZADO ---
    memoria_optimizada = 5 + n + 4

    print(f"Para una lista de N={n} números:")
    print(f"Variables simultáneas en memoria (Fuerza Bruta): ~{int(memoria_fuerza_bruta)}")
    print(f"Variables simultáneas en memoria (Optimizado): ~{int(memoria_optimizada)}")

    mejora = memoria_fuerza_bruta / memoria_optimizada
    print(f"\nEl algoritmo optimizado consume unas {int(mejora)} veces menos memoria RAM")

# Lo probamos con los 9 números de nuestro ejemplo
complejidad_espacial_comparacion(9)
````
